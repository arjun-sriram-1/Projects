```python
import sqlite3
```

```python
conn = sqlite3.connect('transactions.db')
cursor = conn.cursor()
```

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS transactions (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    amount REAL,
    oldbalance REAL,
    newbalance REAL,
    prediction TEXT
)
''')

conn.commit()
print("Table created!")
```
