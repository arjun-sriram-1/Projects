```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import joblib
```

```python
data = pd.read_csv('../data/transactions.csv')

X = data[['amount', 'oldbalance', 'newbalance']]
y = data['is_fraud']
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
joblib.dump(model, 'fraud_model.pkl')
print("Model saved!")
```
